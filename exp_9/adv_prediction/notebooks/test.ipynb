# Testing model

```python
model = SimpleMultiLabelPrediction(input_dim=768,
                                   labels=dataset.unique_labels,
                                   device="cuda")
```

```python
model.device
```

```python
model = model.to("cuda")
```

```python
criterion = nn.BCELoss(reduction='sum')
```

```python
optimizer = optim.Adam(model.parameters(), lr =3e-3)
```

```python
data = torch.rand(16, 768)
targets = torch.rand(16, 36)
data = data.to("cuda")
targets = targets.to("cuda")
```

```python
for i in enumerate(range(100)):
    y_pred = model(data)
    loss = criterion(y_pred, targets)
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    print(loss.item())
```
